```python
from pathlib import Path

import pandas as pd

base_dir = Path(r"c:\Users\clebe\OneDrive\Documentos\USP\7 semestre\governo-aberto\projeto-gov-aberto")
input_csv = base_dir / "dados" / "escolas_sao_vicente_inep.csv"

df_escolas = pd.read_csv(input_csv, encoding="utf-8-sig")
colunas_analise = [
    "Localização",
    "Categoria Administrativa",
    "Dependência Administrativa",
    "Porte da Escola",
    "Etapas e Modalidade de Ensino Oferecidas",
]

for coluna in colunas_analise:
    print(f"\n=== {coluna} ===")
    contagem = df_escolas[coluna].fillna("Não Informado").astype(str).str.strip().value_counts(dropna=False)
    print(contagem.to_string())
```
