## Dados tratado  de saúde

```python
from pathlib import Path

import pandas as pd

base_dir = Path(r"c:\Users\clebe\OneDrive\Documentos\USP\7 semestre\governo-aberto\projeto-gov-aberto")
input_csv = base_dir / "dados" / "Estabelecimentos_saude_sao_vicente_cnes.xlsx - Planilha1.csv"
output_dir = base_dir / "dados" / "dados-tratados"
output_csv = output_dir / "estabelecimentos_saude_sao_vicente_cnes_tratado.csv"

colunas_saida = [
    "NO_FANTASIA",
    "DS_ESFERA_ADMINISTRATIVA",
    "CO_CEP",
    "NO_LOGRADOURO",
    "NU_ENDERECO",
    "NO_BAIRRO",
    "NU_TELEFONE",
    "NU_LATITUDE",
    "NU_LONGITUDE",
    "DS_TURNO_ATENDIMENTO",
    "ST_CENTRO_CIRURGICO",
    "ST_CENTRO_OBSTETRICO",
    "ST_CENTRO_NEONATAL",
    "ST_ATEND_HOSPITALAR",
    "ST_SERVICO_APOIO",
    "ST_ATEND_AMBULATORIAL",
    "CO_AMBULATORIAL_SUS",
]

novos_nomes_colunas = {
    "NO_FANTASIA": "nome-fantasia",
    "DS_ESFERA_ADMINISTRATIVA": "esfera-administrativa",
    "CO_CEP": "cep",
    "NO_LOGRADOURO": "logradouro",
    "NU_ENDERECO": "numero",
    "NO_BAIRRO": "bairro",
    "NU_TELEFONE": "telefone",
    "NU_LATITUDE": "latitude",
    "NU_LONGITUDE": "longitude",
    "DS_TURNO_ATENDIMENTO": "turno-atendimento",
    "ST_CENTRO_CIRURGICO": "centro-cirurgico",
    "ST_CENTRO_OBSTETRICO": "centro-obstetrico",
    "ST_CENTRO_NEONATAL": "centro-neonatal",
    "ST_ATEND_HOSPITALAR": "atendimento-hospitalar",
    "ST_SERVICO_APOIO": "servico-apoio",
    "ST_ATEND_AMBULATORIAL": "atendimento-ambulatorial",
    "CO_AMBULATORIAL_SUS": "ambulatorial-sus",
}

mapeamento_turno = {
    "ATENDIMENTOS NOS TURNOS DA MANHA E A TARDE": "MANHA-TARDE",
    "ATENDIMENTO NOS TURNOS DA MANHA, TARDE E NOITE": "MANHA-TARDE-NOITE",
    "ATENDIMENTO COM TURNOS INTERMITENTES": "INTERMITENTES",
    "ATENDIMENTO CONTINUO DE 24 HORAS/DIA (PLANTAO:INCLUI SABADOS, DOMINGOS E FERIADOS)": "24H",
    "ATENDIMENTO SOMENTE A TARDE": "TARDE",
    "ATENDIMENTO SOMENTE PELA MANHA": "MANHA",
}

output_dir.mkdir(parents=True, exist_ok=True)

df_saude = pd.read_csv(input_csv, encoding="utf-8-sig")
df_saida = df_saude[colunas_saida].copy()
df_saida["DS_TURNO_ATENDIMENTO"] = (
    df_saida["DS_TURNO_ATENDIMENTO"]
    .fillna("Não Informado")
    .astype(str)
    .str.strip()
    .replace(mapeamento_turno)
)

df_saida = df_saida.rename(columns=novos_nomes_colunas)
df_saida.to_csv(output_csv, index=False, encoding="utf-8-sig")

print(f"Arquivo gerado: {output_csv}")
print(df_saida.head())
```
