## Dados tratados Educacao

```python
from pathlib import Path

import pandas as pd

base_dir = Path(r"c:\Users\clebe\OneDrive\Documentos\USP\7 semestre\governo-aberto\projeto-gov-aberto")
input_csv = base_dir / "dados" / "escolas_sao_vicente_inep.csv"
output_dir = base_dir / "dados" / "dados-tratados"
output_csv = output_dir / "estabelecimentos_educacao_sao_vicente_inep_tratado.csv"

colunas_saida = [
    "Restrição de Atendimento",
    "Escola",
    "UF",
    "Município",
    "Categoria Administrativa",
    "Endereço",
    "Telefone",
    "Dependência Administrativa",
    "Etapas e Modalidade de Ensino Oferecidas",
    "Latitude",
    "Longitude",
]

novos_nomes_colunas = {
    "Restrição de Atendimento": "restricao-atendimento",
    "Escola": "escola",
    "UF": "uf",
    "Município": "municipio",
    "Categoria Administrativa": "categoria-administrativa",
    "Endereço": "endereco",
    "Telefone": "telefone",
    "Dependência Administrativa": "dependencia-administrativa",
    "Etapas e Modalidade de Ensino Oferecidas": "etapas-modalidade-ensino",
    "Latitude": "latitude",
    "Longitude": "longitude",
}

mapeamento_etapas = {
    "Educação Infantil": "INFANTIL",
    "Ensino Fundamental": "FUNDAMENTAL",
    "Ensino Médio": "MEDIO",
    "Educação de Jovens e Adultos": "EJA",
    "Educação Profissional": "PROFISSIONAL",
    "Educação Especial": "ESPECIAL",
}


def normalizar_etapas(valor):
    if pd.isna(valor):
        return "NAO INFORMADO"

    texto = str(valor).strip()
    if not texto:
        return "NAO INFORMADO"

    partes = [parte.strip() for parte in texto.split(",") if parte.strip()]
    partes_mapeadas = [mapeamento_etapas.get(parte, parte.upper()) for parte in partes]
    return "-".join(partes_mapeadas)


output_dir.mkdir(parents=True, exist_ok=True)

df_educacao = pd.read_csv(input_csv, encoding="utf-8-sig")
df_saida = df_educacao[colunas_saida].copy()
df_saida["Etapas e Modalidade de Ensino Oferecidas"] = (
    df_saida["Etapas e Modalidade de Ensino Oferecidas"].apply(normalizar_etapas)
)

df_saida = df_saida.rename(columns=novos_nomes_colunas)
df_saida.to_csv(output_csv, index=False, encoding="utf-8-sig")

print(f"Arquivo gerado: {output_csv}")
print(df_saida.head())
```
